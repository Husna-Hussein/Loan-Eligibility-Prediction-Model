## LOAN ELIGIBILITY PREDICTION MODEL USING RANDOM FOREST CLASSIFIER

## Step 1: Importing the dependencies

```python
# importing the required libraries
import pandas as pd # reading and manipulating the data
import numpy as np

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score

import pickle as pkl
```

## Step 2: Load the datasets

```python
# read the data to a pandas Dataframe
train_data = pd.read_csv("data/train.csv")
test_data = pd.read_csv("data/test.csv")
```

```python
# displaying the first five rows of the train data
train_data.head()
```

```python
# displaying the first five rows of the test data
test_data.head()
```

```python
train_data.shape
```

```python
train_data.info()
```

## Step 3: Data Preprocessing

```python
# check for missing data
train_data.isnull().sum()
```

```python
# Checking the statistical info
train_data.describe()
```

```python
train_data["Dependents"].replace({"3+":4},inplace=True)
train_data["Dependents"]
```

```python
test_data["Dependents"].replace({"3+":4},inplace=True)
test_data["Dependents"].value_counts()
```

```python
# analysing distribution of the columns with missing values
# Credit_History column
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(train_data.Credit_History)
```

```python
# LoanAmount column
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(train_data.LoanAmount)
```

```python
# Loan_Amount_Term column
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(train_data.Loan_Amount_Term)
```

## Handling the missing values

```python
# checking the value counts
train_data['Loan_Amount_Term'].value_counts()
```

```python
# checking the value counts
train_data['Self_Employed'].value_counts()
```

```python
# checking the value counts
train_data['Gender'].value_counts()
```

```python
train_data['Gender'].fillna(train_data['Gender'].mode()[0], inplace=True)
test_data['Gender'].fillna(test_data['Gender'].mode()[0], inplace=True)

train_data['Dependents'].fillna(train_data['Dependents'].mode()[0], inplace=True)
test_data['Dependents'].fillna(test_data['Dependents'].mode()[0], inplace=True)

train_data['Married'].fillna(train_data['Married'].mode()[0], inplace=True)
test_data['Married'].fillna(test_data['Married'].mode()[0], inplace=True)

train_data['Credit_History'].fillna(train_data['Credit_History'].mean(), inplace=True)
test_data['Credit_History'].fillna(test_data['Credit_History'].mean(), inplace=True)

train_data['LoanAmount'].fillna(train_data['LoanAmount'].mean(), inplace=True)
test_data['LoanAmount'].fillna(test_data['LoanAmount'].mean(), inplace=True)

train_data['Self_Employed'].fillna(train_data['Self_Employed'].mode()[0], inplace=True)
test_data['Self_Employed'].fillna(test_data['Self_Employed'].mode()[0], inplace=True)

train_data['Loan_Amount_Term'].fillna(train_data['Loan_Amount_Term'].mode()[0], inplace=True)
test_data['Loan_Amount_Term'].fillna(test_data['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
# check for missing data
train_data.isnull().sum()
```

## Label Encoding

```python
# converting rows with string values into numerical numbers


#Gender,Married,Education,Self_Employed,Property_Area
train_data["Gender"].replace({"Male":0,"Female":1},inplace=True)
train_data["Married"].replace({"No":0,"Yes":1},inplace=True)
train_data["Education"].replace({"Not Graduate":0,"Graduate":1},inplace=True)
train_data["Self_Employed"].replace({"No":0,"Yes":1},inplace=True)
train_data["Property_Area"].replace({"Rural":0,"Urban":1,"Semiurban":2},inplace=True)
train_data["Loan_Status"].replace({"N":0,"Y":1},inplace=True)

train_data
```

```python
# encoding the test data
test_data["Gender"].replace({"Male":0,"Female":1},inplace=True)
test_data["Married"].replace({"No":0,"Yes":1},inplace=True)
test_data["Education"].replace({"Not Graduate":0,"Graduate":1},inplace=True)
test_data["Self_Employed"].replace({"No":0,"Yes":1},inplace=True)
test_data["Property_Area"].replace({"Rural":0,"Urban":1,"Semiurban":2},inplace=True)
```

```python
test_data
```

## Plot the correlation

```python
plt.figure(figsize=(12,8))
cmap = sns.diverging_palette(260,-10,s=50, l=75, n=6, as_cmap=True)
sns.heatmap(train_data.corr(), cmap=cmap, annot=True, fmt='.1f', linewidths=.1)
plt.show()
```

## Step 4: Training the Random Forest Classifier

```python
X_train =train_data.drop(train_data[["Loan_ID","Loan_Status"]],axis=1)
X_train
```

```python
y_train=  train_data["Loan_Status"]
y_train
```

## Normalize the data by Scaling

```python
scaler =StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
# train the model
classifier = RandomForestClassifier(n_estimators=50,max_depth= 9)
#fit function is used to train the model using the training sets as parameters
classifier.fit(X_train, y_train)
```

```python
X_test=test_data[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']]
X_test
```

## Step 5: Evaluating the model

```python
# predict using the model
y_pred_train = classifier.predict(X_train)
y_pred_test = classifier.predict(X_test)
```

```python
# evaluate the model using the training data
acc = accuracy_score(y_train,y_pred_train)
f1= f1_score(y_train,y_pred_train)
prec = precision_score(y_train,y_pred_train)

print(f"Train Accuracy {acc*100:,.2f}%") 
print(f"Train F1 Score {f1*100:,.2f}%" )
print(f"Train Precision Score {prec*100:,.2f}%" )
```

```python
y_pred_test[0]
```

## Step 6: Building a predictive system

```python
y_prediction =classifier.predict([[0,0,0,1,0,5849,0.0,146.412162,360.0,1.0,1]])
y_prediction
```

```python
if y_prediction[0] == 1:
    print("Eligible")
else:
    print("Not Eligible")
    
```

```python
# save the model to disk
filename = 'model.pkl'
pkl.dump(classifier, open(filename, 'wb'))
```
